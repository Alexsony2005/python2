```python
s = input("Enter string: ")

print("Entered String:", s)

# Palindrome check
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

# Symmetrical check
n = len(s)
half = n // 2

first_half = s[:half]
second_half = s[-half:]

if first_half == second_half:
    print("Symmetrical")
else:
    print("Not Symmetrical")
```
